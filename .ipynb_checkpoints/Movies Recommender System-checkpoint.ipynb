## Importing the Libraries

```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

## Laoding the data

```python
movies_rec_sys=pd.read_csv('movies.csv')
```

```python
movies_rec_sys.head()
```

```python
movies_rec_sys.tail()
```

```python
movies_rec_sys.shape
```

```python
selected_features=['genres','keywords','title','cast','director']
print(selected_features)
```

## Filling missing entries in the selected features with null string

```python
for feature in selected_features:
    movies_rec_sys[feature]=movies_rec_sys[feature].fillna('')
```

## Combining the selected features

```python
final_features=movies_rec_sys['genres']+','+movies_rec_sys['keywords']+','+movies_rec_sys['title']+','+movies_rec_sys['cast']+','+movies_rec_sys['director']
```

```python
print(final_features)
```

## Converting textual data to feature vectors

```python
vect=TfidfVectorizer()
vect_features=vect.fit_transform(final_features)
print(vect_features)
```

## Finding the cosine similarity of vectorized features

```python
similarity=cosine_similarity(vect_features)
print(similarity)
```

## Getting the movie name from the user

```python
a=input("Enter the movie name: ")
```

## Creating the list of all movie titles in dataset

```python
titles_list=movies_rec_sys['title'].tolist()
```

```python
print(titles_list)
```

## Finding the close match from the dataset movies with the user movie

```python
related_movie=difflib.get_close_matches(a,titles_list)
print(related_movie)
```

```python
close_match=related_movie[0]
print(close_match)
```

## Searching the index of the user movie in the dataset

```python
user_movie_index=movies_rec_sys[movies_rec_sys.title==close_match]['index'].values[0]
print(user_movie_index)
```

## List of similar movies on the basis of similarity score

```python
sim_score=list(enumerate(similarity[user_movie_index]))
print(sim_score)
```

```python
len(sim_score)
```

## Sorting the similar movies

```python
sorted_movies=sorted(sim_score,key= lambda x:x[1],reverse=True)
print(sorted_movies)
```

## Suggesting the familiar movies via index

```python
print("Your Movie Recommendations are:\n")

i=1
for movie in sorted_movies:
    index=movie[0]
    movie_title=movies_rec_sys[movies_rec_sys.index==index]['title'].values[0]
    if (i<10):
        print(i,':',movie_title)
        i+=1
```

## Creating a Pipeline

```python
a=input("Enter the movie name: ")
titles_list=movies_rec_sys['title'].tolist()
related_movie=difflib.get_close_matches(a,titles_list)
close_match=related_movie[0]
user_movie_index=movies_rec_sys[movies_rec_sys.title==close_match]['index'].values[0]
sim_score=list(enumerate(similarity[user_movie_index]))
sorted_movies=sorted(sim_score,key= lambda x:x[1],reverse=True)
print("Your Movie Recommendations are:\n")

i=1
for movie in sorted_movies:
    index=movie[0]
    movie_title=movies_rec_sys[movies_rec_sys.index==index]['title'].values[0]
    if (i<10):
        print(i,':',movie_title)
        i+=1



```
